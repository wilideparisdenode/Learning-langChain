# Deep Agents: Descripción General

## ¿Qué es Deep Agents?
* `deepagents` es una biblioteca independiente para construir agentes capaces de abordar tareas complejas y de múltiples pasos.
* Construido sobre LangGraph e inspirado en aplicaciones como Claude Code, Deep Research y Manus, los deep agents incluyen:
    * capacidades de planificación,
    * sistemas de archivos para la gestión del contexto,
    * y la capacidad de crear subagentes.

## ¿Cuándo usar Deep Agents?
* Utiliza deep agents cuando necesites agentes que puedan:
    * Manejar tareas complejas y de múltiples pasos que requieran planificación y descomposición.
        * Dividir tareas complejas en pasos discretos.
    * Gestionar grandes cantidades de contexto mediante herramientas del sistema de archivos.
    * Delegar trabajo a subagentes especializados para el aislamiento del contexto.
    * **Persistir la memoria entre conversaciones e hilos**.
        * Ampliar los agentes con memoria persistente entre hilos **utilizando el Store de LangGraph**.
* Para casos de uso más sencillos, considera usar create_agent de LangChain o construir un flujo de trabajo personalizado con LangGraph.

## Relación con el Ecosistema de LangChain
* Deep agents está construido sobre:
    * **LangGraph** - Proporciona la **ejecución del grafo subyacente y la gestión del estado**.
    * **LangChain - Las herramientas e integraciones de modelos** funcionan sin problemas con deep agents.
    * **LangSmith - Observabilidad, evaluación y despliegue**.
