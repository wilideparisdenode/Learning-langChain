# LangChain 1.0 Agente Post-Despliegue: Observabilidad con LangSmith
* Los agentes de LangChain admiten automáticamente el trazado a través de LangSmith para capturar, depurar, evaluar y monitorizar el comportamiento.
* Las trazas registran cada paso de la ejecución del agente, desde la entrada inicial del usuario hasta la respuesta final, incluyendo todas las llamadas a herramientas, interacciones con el modelo y puntos de decisión.
* Estos datos de ejecución te ayudan a depurar problemas, evaluar el rendimiento en diferentes entradas y monitorizar patrones de uso en producción.
* Para ello, necesitarás:
    * Una cuenta de LangSmith (registro gratuito).
    * Una clave API de LangSmith.
* Para aprender más sobre cómo usar LangSmith, consulta la [documentación](https://docs.langchain.com/oss/python/langchain/observability).
