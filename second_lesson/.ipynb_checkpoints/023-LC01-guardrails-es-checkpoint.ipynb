# LangChain 1.0 Temas Avanzados: Guardrails

## ¿Qué son los Guardrails y por qué son importantes?
* Los Guardrails te ayudan a construir aplicaciones de IA seguras y conformes validando y filtrando contenido en puntos clave de la ejecución de tu agente.
* Los casos de uso comunes incluyen:
    * Prevenir la fuga de PII (Información Privada).
    * Detectar y bloquear ataques de inyección de prompts.
    * Bloquear contenido inapropiado o dañino.
    * Hacer cumplir reglas de negocio y requisitos de conformidad.
    * Validar la calidad y precisión de la salida.

## Hay dos tipos principales de Guardrails
* Guardrails Manuales: utilizando lógica basada en reglas como patrones regex, coincidencia de palabras clave o comprobaciones explícitas.
* Guardrails basados en Modelos: utilizando LLMs para evaluar contenido.
    * Estos suelen tener mayor calidad, pero son más lentos y costosos.

## ¿Cómo implementamos Guardrails en LangChain 1.0?
* Puedes implementar guardrails utilizando middleware.
* LangChain 1.0 proporciona tanto guardrails integrados (por ejemplo, detección de PII, human-in-the-loop) como un sistema de middleware flexible para construir guardrails personalizados utilizando cualquiera de los dos enfoques.
* Consulta la [documentación](https://docs.langchain.com/oss/python/langchain/guardrails) para obtener más información al respecto.
