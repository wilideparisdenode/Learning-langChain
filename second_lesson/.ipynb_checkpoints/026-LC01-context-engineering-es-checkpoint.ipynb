# LangChain 1.0 Temas Avanzados: Ingeniería de Contexto

## La principal causa de agentes de baja calidad: no se pasó el contexto adecuado al Modelo
* Las aplicaciones LLM y los agentes de IA fallan por una de estas dos razones:
    * El LLM subyacente no es lo suficientemente capaz.
    * El contexto "adecuado" no se pasó al LLM.
* Más a menudo de lo que parece - en realidad es la segunda razón la que hace que los agentes no sean fiables.

## ¿Qué es la Ingeniería de Contexto?
* La Ingeniería de Contexto consiste en proporcionar la información y las herramientas adecuadas en el formato correcto para que el Modelo pueda realizar una tarea.

#### Para construir agentes fiables, necesitas controlar lo que sucede en cada paso del bucle del agente, así como lo que ocurre entre pasos:
* Contexto del modelo: Qué entra en las llamadas al modelo (instrucciones, historial de mensajes, herramientas, formato de respuesta).
* Contexto de las herramientas: A qué pueden acceder y producir las herramientas (lecturas/escrituras en estado, almacén, contexto de ejecución).
* Contexto del ciclo de vida: Qué ocurre entre las llamadas al modelo y a las herramientas (resumen, barreras de seguridad, registro, etc.).

#### En la Ingeniería de Contexto, puedes acceder a tres fuentes de datos:
* Contexto de ejecución (también conocido como configuración estática).
    * ID de usuario,
    * claves API,
    * conexiones a bases de datos,
    * permisos,
    * configuración del entorno, etc.
* Estado (también conocido como memoria a corto plazo).
    * Mensajes actuales,
    * archivos subidos,
    * estado de autenticación,
    * resultados de herramientas, etc.
* Almacén (también conocido como memoria a largo plazo).
    * Preferencias del usuario,
    * conocimientos extraídos,
    * recuerdos,
    * datos históricos, etc.

## ¿Cómo se implementa la Ingeniería de Contexto en LangChain 1.0?
* Utilizando middleware.
* El middleware te permite conectarte a cualquier paso del ciclo de vida del agente y:
    * Actualizar el contexto.
    * Saltar a un paso diferente en el ciclo de vida del agente.
* Consulta la [documentación](https://docs.langchain.com/oss/python/langchain/context-engineering) para más información al respecto.
