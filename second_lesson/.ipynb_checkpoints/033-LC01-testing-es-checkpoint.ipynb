# Desarrollo de Agentes con LangChain 1.0: Testing
* Existen varios enfoques para hacer pruebas de tus agentes:
    * Las pruebas unitarias prueban pequeñas piezas deterministas de tu agente de forma aislada utilizando simulaciones en memoria, de modo que puedes verificar el comportamiento exacto de manera rápida y determinista.
    * Las pruebas de integración prueban el agente utilizando llamadas reales de red para confirmar que los componentes funcionan juntos, que las credenciales y esquemas coinciden, y que la latencia es aceptable.
* Las aplicaciones agénticas tienden a depender más de las pruebas de integración porque encadenan múltiples componentes y deben lidiar con la inestabilidad debido a la naturaleza no determinista de los LLMs.
* Para aprender más sobre cómo hacer pruebas, consulta la [documentación](https://docs.langchain.com/oss/python/langchain/test).
