# Despliegue de Agentes LangChain 1.0: Despliegue con LangSmith
* LangSmith proporciona una plataforma de alojamiento gestionada diseñada para cargas de trabajo de agentes.
* Las plataformas de alojamiento tradicionales están diseñadas para aplicaciones web sin estado y de corta duración, mientras que LangGraph está especialmente diseñado para agentes con estado y de larga ejecución que requieren estado persistente y ejecución en segundo plano.
* LangSmith gestiona la infraestructura, el escalado y las preocupaciones operativas para que puedas desplegar directamente desde tu repositorio.
* Para poder hacerlo, necesitarás:
    * Una cuenta de GitHub.
    * Una cuenta de LangSmith (registro gratuito).
* Para obtener más información sobre cómo desplegar tu agente con LangSmith, consulta la [documentación](https://docs.langchain.com/oss/python/langchain/deploy).
