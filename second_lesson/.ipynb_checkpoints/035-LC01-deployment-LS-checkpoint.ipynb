# LangChain 1.0 Agent Deployment: Deployment with LangSmith
* LangSmith provides a managed hosting platform designed for agent workloads.
* Traditional hosting platforms are built for stateless, short-lived web applications, while LangGraph is purpose-built for stateful, long-running agents that require persistent state and background execution.
* LangSmith handles the infrastructure, scaling, and operational concerns so you can deploy directly from your repository.
* In order to do that, you will need :
    * A GitHub account.
    * A LangSmith account (free to sign up).
* To learn more about how to deploy your agent with LangSmith, see the [documentation](https://docs.langchain.com/oss/python/langchain/deploy).
