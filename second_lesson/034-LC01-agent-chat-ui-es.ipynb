# Desarrollo de Agentes con LangChain 1.0: Interfaz de Chat para Agentes
* Agent Chat UI es una aplicación Next.js que proporciona una interfaz conversacional para interactuar con cualquier agente de LangChain.
* Es compatible con:
    * chat en tiempo real,
    * visualización de herramientas,
    * y funcionalidades avanzadas como la depuración con viaje en el tiempo
    * y la bifurcación de estados.
* Agent Chat UI funciona perfectamente con agentes creados mediante create_agent y proporciona experiencias interactivas para tus agentes con una configuración mínima, tanto si se ejecutan localmente como en un contexto desplegado (como LangSmith).
* Consulta nuestra lección detallada en la Quickstart de LangChain 1.0 para aprender cómo utilizar Agent Chat UI.
