# LangChain 1.0 Advanced Topics: Guardrails

## What are Guardrails and why are they important?
* Guardrails help you build safe, compliant AI applications by validating and filtering content at key points in your agentâ€™s execution.
* Common use cases include:
    * Preventing PII (Private Information) leakage.
    * Detecting and blocking prompt injection attacks.
    * Blocking inappropriate or harmful content.
    * Enforcing business rules and compliance requirements.
    * Validating output quality and accuracy.

## There are two main types of Guardrails
* Manual Guardails: using rule-based logic like regex patterns, keyword matching, or explicit checks.
* Model-based Guardrails: using LLMs to evaluate content.
    * These use to have higher quality, but are slower and more expensive.

## How do we implement Guardrails in LangChain 1.0?
* You can implement guardrails using middleware.
* LangChain 1.0 provides both built-in guardrails (e.g., PII detection, human-in-the-loop) and a flexible middleware system for building custom guardrails using either approach.
* See the [documentation](https://docs.langchain.com/oss/python/langchain/guardrails) for more info about this.
